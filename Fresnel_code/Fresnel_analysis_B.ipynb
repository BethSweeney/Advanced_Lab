```python
def theoretical_straight_edge(z1,z2,yy):
    Y = (z1/(z2+z1))*yy
    V1 = Y*(np.sqrt((2*(z1+z2))/((z1)*(340/(40000*2*np.pi)))))
    S,C = sc.fresnel(V1)
    
    ans = 1/4*(np.abs(1+1j)**2*np.abs(1/2*(1+1j)-C-1j*S)**2)
    return ans,C,S
```
